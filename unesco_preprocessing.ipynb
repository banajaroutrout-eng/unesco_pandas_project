 Drop columns with too many missing values

```python
c_df=c_df.drop(columns=['unique_number','id_no','rev_bis','short_description_en','secondary_dates','date_end','danger_list','criteria_txt','iso_code','udnp_code','transboundary','category_short'])
```

```python
c_df.columns
```

```python
c_df = c_df.drop('justification_en', axis=1)
```

Change datatypes

```python
# Fill text columns with 'Unknown'
c_df['states_name_en'] = c_df['states_name_en'].fillna('Unknown')

c_df['region_en'] = c_df['region_en'].fillna('Unknown')

c_df['category'] = c_df['category'].fillna('Unknown')

c_df['name_en'] = c_df['name_en'].fillna('Unknown')
```

Convert year column to numeric

```python
c_df['date_inscribed'] = c_df['date_inscribed'].fillna(0)

c_df['longitude'] = c_df['longitude'].fillna(0)

c_df['latitude'] = c_df['latitude'].fillna(0)

c_df['area_hectares'] = c_df['area_hectares'].fillna(0)
```

```python
c_df.isnull().sum()
```

```python
c_df.head()
```

```python
c_df.info()
```

Rename Columns

```python
c_df = c_df.rename(columns={
    'category': 'Category',
    'states_name_en': 'Country',
    'region_en': 'Region',
    'name_en': 'Site_Name',
    'date_inscribed': 'Year',
    'danger': 'Danger',
    'longitude': 'Longitude',
    'latitude': 'Latitude',
    'area_hectares': 'Area'
})
```

```python
c_df.head()
```

```python
c_df.info()
```

# Feature Engineering (Data transformation)

Create age of site column

```python
c_df['Age']=2026-c_df['Year']  # It calculates how many years have passed since the site was added
```

```python
c_df.head()
```

Create area category column

```python

c_df['Area_Category'] = pd.cut(
    c_df['Area'],
    bins=[-1, 10000, 100000, c_df['Area'].max()],
    labels=['Small', 'Medium', 'Large']
)
```

Create danger status column

```python

c_df['Danger_Status'] = c_df['Danger'].map({ 0: 'Safe',1: 'In Danger'}) # Numeric danger values are converted into readable labels.
```

Extract decade from year

```python
c_df['Decade'] = (c_df['Year'] // 10) * 10   # In this step, the decade of inscription is created from the Year column.
```

```python
c_df.head()
```

```python
c_df = c_df.sort_values(by='Year')  #the dataset is sorted based on Year.
```

```python
c_df = c_df.reset_index(drop=True)
```

```python
c_df.to_csv('unesco_cleaned.csv', index=False)

print("File saved successfully")
```
