```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
c_df=pd.read_csv('unesco_cleaned.csv')
```

```python
c_df.columns
```

# 1.How many heritage sites are there in each category?

```python
plt.figure(figsize=(5,2))
sns.countplot(c_df,x='Category')
plt.title('Category wise no_of_sites')
plt.show()
```

# Observation

There are 3 heritage sites

There are about 870 cultural sites, 210 natural sites, and 40 mixed sites

# 2.Which countries have the highest number of heritsge sites?

```python
plt.figure(figsize=(6,2))
sns.countplot(c_df,y='Region')
plt.title('Region_wise no_of sites')
plt.ylabel('no_of_sites')
plt.yticks(fontsize=10)
plt.show()
```

# Observation

Countries in Europe and North America have the highest number of heritage sites.

# 3.How are heritage sites distributed across Regions?

```python
c_df['Region'].value_counts().plot.pie(autopct='%1.1f%%') #automatic percentage
plt.title('Region-wise Distribution of Heritage Sites')
plt.ylabel('')
plt.show()
```

Europe & North America has the highest number of heritage sites (almost half).

Asia & Pacific is the second highest region.

Latin America & Caribbean has a moderate share.

Africa and Arab States have the least number of sites

# 4.How many sites were added each year?

```python
year_counts = c_df['Year'].value_counts().sort_index()

plt.figure(figsize=(8,3))

plt.plot(year_counts.index,
         year_counts.values)

plt.title('year wise no_of_sites')
plt.xlabel('Year')
plt.ylabel('no_of_sites')

plt.xticks(fontsize=6)

plt.show()
```

# Observation
The number of sites added varies every year (not constant).

There is a peak around the year 2000, where the highest number of sites were added.

Some years (like late 1980s and early 2000s) show a sharp drop in additions.

After 2000, the number of sites added becomes more stable but slightly lower.

Overall, there is a fluctuating trend with no steady increase or decrease.

# 5.How many sites fall into each Area category?

```python
plt.figure(figsize=(6,3))

sns.countplot(c_df, x='Area_Category')

plt.title('area_category wise no_of_sites')
plt.xlabel('Area_Category')
plt.ylabel('no_of_sites')

plt.xticks(fontsize=7)

plt.show()
```

# Observation
Small area sites are the highest in number.

Large area sites are more than medium, but much less than small.

Medium area sites are the least.


#  6.How many heritage sites are in danger vs safe?

```python
plt.figure(figsize=(5,3))

sns.countplot(c_df, x='Danger_Status')

plt.title('danger_status wise no_of_sites')
plt.ylabel('no_of_sites')
plt.xlabel('Danger_Status')

plt.xticks(fontsize=8)

plt.show()
```

# Observation

The majority of heritage sites are safe with very few classified as danger


# 7. Which regions have the most endangered heritage site?

```python
plt.figure(figsize=(3,2))

sns.countplot(data=c_df,
              y='Region',
              hue='Danger_Status')

plt.title('region vs danger_status')
plt.xlabel('Region')
plt.ylabel('no_of_sites')

plt.xticks(fontsize=5,rotation=90)

plt.tight_layout()

plt.show()
```

# Observation

Latin America & the Caribbean and Africa have the highest number of endangered heritage sites compared to other regions.

# 8. What is the age distribution of site ages?

```python
plt.figure(figsize=(7,3))

sns.histplot(data=c_df,
             x='Age')

plt.title('age distribution of heritage sites')
plt.xlabel('Age')
plt.ylabel('frequency')

plt.show()
```

# Observation

Most heritage sites are in the mid-age range(around 20-40 years) with fewer very young and very old sites

# 9. What is the relationship between site Area and Age?

```python
plt.figure(figsize=(7,3))

sns.scatterplot(data=c_df,
                x='Age',
                y='Area')

plt.title('age vs area of heritage sites')
plt.xlabel('Age')
plt.ylabel('Area')

plt.show()
```

# Observations
There is no clear relationship between age and area.

Most sites, regardless of age, have small area values.

A few sites have very large areas (outliers).

These large-area sites are scattered across different ages, not concentrated in one age group.
